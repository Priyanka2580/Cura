```python
# CELL 1: Setup
import os
import sys

# PaddleX defaults to MKLDNN for CPU inference (run_mode="mkldnn"), which
# enables oneDNN in the paddle inference Config and crashes in PIR with:
#   NotImplementedError: ConvertPirAttribute2RuntimeAttribute not support
#   [pir::ArrayAttribute<pir::DoubleAttribute>]
# Setting this to "0" forces run_mode="paddle" (plain CPU, no oneDNN).
# Must be set before paddleocr / paddlex is first imported in Cell 3.
os.environ["PADDLE_PDX_ENABLE_MKLDNN_BYDEFAULT"] = "0"

sys.path.insert(0, ".")

test_image = r'c:\Users\PRIYANKA\Desktop\MedAi\test_images\027img.png'
print(f"Testing with: {test_image}")
```

```python
# CELL 2: Test Classifier
from models.image_classifier import DocumentClassifier

clf = DocumentClassifier()
result = clf.predict(test_image)
print(result)
```

```python
# CELL 3: Test OCR
from models.ocr_engine import OCREngine

ocr = OCREngine()
text = ocr.extract_text(test_image)
print(text["raw_text"][:500])
```

```python
# CELL 4: Test NER
from models.ner_extractor import NERExtractor

ner = NERExtractor()
entities = ner.extract_entities(text["raw_text"])
print(entities)
```
